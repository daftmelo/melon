```python
matplotlib实际上是一种比较低级的工具。
要绘制一张图表，你组装一些基本组件就行：数据展示（即图表类型：线型图、柱状图、盒形图、散布图、等值线图等）、图例、标题、刻度标签以及其他注解型信息。
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
@线型图
```

```python
Series和DataFrame都有一个用于生成各类图表的plot方法。
默认情况下，它们所生成的是线型图：
```

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
```

```python
# 简单的Series图标示例：
s.plot()
```

```python
s
```

```python
该Series对象的索引会被传给matplotlib，并用以绘制X轴。可以通过use_index=False禁用该功能。（不过你为什么要这么做呢）
X轴的"刻度"和"界限"可以通过xticks和xlim选项进行调节，Y轴就用yticks和ylim。
```

```python
DataFrame的plot方法会在一个subplot中为"各列"绘制一条线，并自动创建图例：
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
```

```python
# 简单的DataFrame绘图：
df.plot()
```

```python
plot属性包含一批不同绘图类型的方法。（kind参数可供选择）
例如，df.plot()等价于df.plot.line()。后面会学习这些方法。
```

```python
笔记：plot的其他关键字参数会被传给相应的matplotlib绘图函数，所以要更深入地自定义图表，就必须学习更多有关matplotlib API的知识。
```

```python
DataFrame还有一些用于对列进行灵活处理的选项。
例如，是要将所有列都绘制到一个subplot中还是创建各自的subplot。
```
