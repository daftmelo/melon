```python
matplotlib实际上是一种比较低级的工具。
要绘制一张图表，你组装一些基本组件就行：数据展示（即图表类型：线型图、柱状图、盒形图、散布图、等值线图等）、图例、标题、刻度标签以及其他注解型信息。
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
@线型图
```

```python
Series和DataFrame都有一个用于生成各类图表的plot方法。
默认情况下，它们所生成的是线型图：
```

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
```

```python
# 简单的Series图标示例：
s.plot()
```

```python
s
```

```python
该Series对象的索引会被传给matplotlib，并用以绘制X轴。可以通过use_index=False禁用该功能。（不过你为什么要这么做呢）
X轴的"刻度"和"界限"可以通过xticks和xlim选项进行调节，Y轴就用yticks和ylim。
```

```python
pandas的大部分绘图方法都有一个可选的ax参数，它可以是一个matplotlib的subplot对象。这使你能够在网格布局中更为灵活地处理subplot的位置。
```

```python
DataFrame的plot方法会在一个subplot中为"各列"绘制一条线，并自动创建图例：
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
```

```python
# 简单的DataFrame绘图：
df.plot() # default line
```

```python
plot属性包含一批不同绘图类型的方法。（kind参数可供选择）
例如，df.plot()等价于df.plot.line()。后面会学习这些方法。
```

```python
笔记：plot的其他关键字参数会被传给相应的matplotlib绘图函数，所以要更深入地自定义图表，就必须学习更多有关matplotlib API的知识。
```

```python
DataFrame还有一些用于对列进行灵活处理的选项。
例如，是要将所有列都绘制到一个subplot中还是创建各自的subplot。
```

```python
参数            说明
---------------------
subplots       将各个DataFrame列绘制到单独的subplot中
sharex         如果subplots=True，则共用一个X轴，包括刻度和界限
sharey         如果subplots=True，则共用一个Y轴
legend         添加一个subplot图例（默认为True）
sort_columns   以字母表顺序绘制各列，默认使用当前列顺序
```

```python
@柱状体
```

```python
plot.bar()和plot.barh()分别绘制水平和垂直的柱状图。
这时，Series和DataFrame的索引将会被用作X（bar）或Y（barh）刻度
```

```python
fig, axes = plt.subplots(ncols=1, nrows=2, figsize=(12, 9))
```

```python
data = pd.Series(np.random.randn(16), index=list('abcdefghijklmnop'))
```

```python
data.plot.bar(ax=axes[0], color='k', alpha=0.7)
```

```python
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```

```python
fig
```

```python
color='k'和alpha=0.7设定了图形的颜色为黑色，并使用部分的填充透明度。
```

```python
对于DataFrame，柱状图会将"每一行"的值分为"一组"，并排显示，如图所示：
```

```python
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df
```

```python
df.plot.bar(figsize=(12, 9))
```

```python
注意，DataFrame各列的名称"Genus"被用作了图例的标题。
```

```python
设置stacked=True即可为DataFrame生成堆积柱状图，这样每行的值就会被堆积在一起：
```

```python
# DataFrame的堆积柱状图
df.plot.barh(stacked=True, alpha=0.5, figsize=(12, 9))
```

```python
笔记：柱状图有一个非常不错的用法：利用value_counts图形化显示Series中各值的出现频率，比如s.value_counts().plot.bar()。
```

```python
# 以House Price数据集为例，统计数据集中缺失观察值组数，生成Series：
df_train = pd.read_csv('../my_kaggle_proj/House Prices: Advanced Regression Techniques/data/train.csv')
```

```python
# 数据集的81个变量中，统计缺失值（miss_data）的观察（observe）数的组数出现的频率
df_train.isnull().sum().value_counts().sort_index().plot.bar()
```
